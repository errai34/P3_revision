# Small code to split the original data into training and testing sets

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

np.random.seed(1234)
```

```python
data = pd.read_csv('../train_data/AllOriginal.csv')
print("Length of data before dropping NaN: ", len(data))
```

```python
data.describe()
```

```python
data = data.dropna()
print("Length of data after dropping NaN: ", len(data))
```

```python
data.head()
```

```python
data_train, data_test = train_test_split(data, test_size = 0.20, random_state = 42)
```

```python
data_train.head()
```

```python
len(data_train)
```

```python
len(data_test)
```

```python
data_train.to_csv('../train_data/TrainingOriginal.csv')
data_test.to_csv('../test_data/TestOriginal.csv')
```
